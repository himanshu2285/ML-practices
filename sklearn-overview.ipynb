# Scikit-learn Theory

### 1. Flow: Every scikit-learn model follows the same pattern:

from sklearn.some_module import SomeModel

Instantiate the model:<br>
model = SomeModel(hyperparameter1 = value1, hyperparameter2 = value2)

Train the model:<br>
model.fit(x_train, y_train)

Make Predictions:<br>
prediction = model.predict(x_test)

Evaluate (For Supervised Learning):<br>
score = model.score(x_test, y_test)
