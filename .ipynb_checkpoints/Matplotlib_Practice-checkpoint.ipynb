```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_points = np.array([5,8])
y_points = np.array([1,5])
plt.plot(x_points, y_points)
plt.show()
```

Plot a line chart

```python
ypoints = np.array([2,6,1,9,4])
plt.plot(ypoints, marker="o",linestyle="dotted")
plt.show()
```

```python
x = np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.plot(x,y, marker="o", linestyle="solid")
plt.show()
```

```python
x = np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
```

### Bar Chart

```python
x=np.array(["A","B","C","D"])
y=np.array([2,7,11,5])
plt.bar(x,y)
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(175, 10, 250) # (mean, sd, 250 random numbers)
plt.hist(x)
plt.show()
```

```python
x = np.array([5,45,25,15,10])
plt.pie(x)
plt.show()
```
