#### Map Function
map(function, iterable)  
function perform on all the items of Iterable(list, tuple, set)

```python
def square(x):
    return x*x
```

```python
lst = np.array([1,2,3,4,5])
result = map(square, lst)
```

```python
result_list = list(result)
result_list
```

### Statistical Function (sum, mean, median, std, var, min, max, argmin, argmax)

```python
np.sum(result_list)
```

```python
np.mean(result_list)
```

```python
np.median(result_list)
```

```python
np.std(result_list)
```

```python
np.var(result_list)
```

```python
np.argmin(result_list)
```

```python
np.argmax(result_list)
```
