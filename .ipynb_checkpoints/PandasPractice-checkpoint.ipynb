#### Know how to add Column names and row names(index)

```python
# Your code goes here. Create a dataframe matching the above diagram and assign it to the variable fruit_sales.
import numpy as np
import pandas as pd
fruit_sales = np.array([[35,21],[41,34]])
fruit_sales_df = pd.DataFrame(fruit_sales)
fruit_sales_df.columns = ["Apples","Bananas"]
fruit_sales_df.index = ["2017 Sales","2018 Sales"]

# Check your answer
fruit_sales_df
# q2.check()
# fruit_sales
```
