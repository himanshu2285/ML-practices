#### prime number

```python
def prime_number(num):
    if num<=1:
        return False
    else:
        for i in range(2, int(num**0.5)+1):
            if num%i==0:
                return False
        return True
    
n = eval(input('Enter a number :'))
prime = prime_number(n)
if prime:
    print(n,'is a prime number')
else:
    print(n,'is not a prime number')
```

#### Factorial

```python
def fact(n: int):
    if n==0 or n==1:
        return 1
    else:
        f=1
        for i in range(1, n+1):
            f*=i
        return f

num = eval(input('Enter a number'))
print(fact(num))
```

#### Sum of prime number

```python
nums = [2,12,13,14,37,59]
def prime_number(num):
    if num<=1:
        return False
    else:
        for i in range(2, int(num**0.5)+1):
            if num%i==0:
                return False
        return True

total=0
for num in nums:
    if prime_number(num):
        total+=num
    else:
        continue
print(total)
```

#### Fibonacci Series

```python
def fiboncci(num: int):
    n1,n2=0,1
    print(f'series : {n1} {n2}', end=' ')
    for i in range(0, num):
        res=n1+n2
        print(res, end=' ')
        n1,n2=n2,res
n=int(input('Enter a number to create fibonacci series :'))
fiboncci(n)
```

#### Rerverse Number

```python
def reverse_num(num):
    num=abs(num)
    rev=0
    while(num>0):
        rem=num%10
        rev=rev*10+rem
        num=num//10
    return rev

try:
    num=int(input(''))
    if num < 0:
        print('-' + str(reverse_num(num)))
    else:
        print(reverse_num(num))
except ValueError:
    print('Given input is not valid')
else:      # it executes when the try block run properly
    print('Our code run properly without any error')
finally:
    print('This block always excute, even if the error occur then also')
    
```
