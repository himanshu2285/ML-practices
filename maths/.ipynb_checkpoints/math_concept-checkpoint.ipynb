### Linear Equation code:

```python
import matplotlib.pyplot as plt
x = [0,1,2,3,4,5]
y=[]
for i in x:
    y.append(2*i+3)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Equation")
plt.show()
```

## Statistics

As a Data Scientist you should know these topics

```python
import numpy as np
arr1 = np.array([10,11,12,12,14,12,17])
range_value = np.max(arr1) - np.min(arr1)
print("Range :",range_value)
```

1. Data Average/Typical value - Mean, Median, Mode
2. Data Spread - How our data is distributed (e.g. A = [10,12,14,16,18] and B = [10,40,150,300,600])
3. Variance - How Far the data is spread from Mean(Average Value)
4. Standerd Deviation - Difference in Mean value and Actual value


mean - if data is consistent then we can use mean e.g. A=[10,20,30,40,50] </br>
median - if data is inconsistent, in that case we use median e.g. A=[10,20,30,40,500,1200] </br>
mode - it find the central tendency based on the most frequent number.

### Variance

Low var = data is less spread from mean <br>
High var = data is more spread

```python
# calculate variance
result = np.var(arr1)
result
```

## Standard Deviation

STD = Sqrt(variance)

```python
data = np.array([10,11,12,12,14,12,17])
print("Mean :",np.mean(data))
std_dev = np.std(data)
print("Standard Deviation :",std_dev)
```

## Data Visualization

```python
# Histogram 
import matplotlib.pyplot as plt
data = [10,20,20,30,30,30,40,50,50]
plt.hist(data)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of marks")
plt.show()
```

```python
# Bar charts
courses = ["AIML", "Analytics","Full stack"]
students = [30,50,20]
plt.bar(courses, students)
plt.xlabel("courses")
plt.ylabel("students")
plt.title("Students for each course")
plt.show()
```
