```python
import numpy as np
```

```python
# Simple List Example in python
l1 = [1,3,4,6,7]
l2 = [3,2,5,6,7]
print(l1 + l2)
```

```python
# Vector Addition and Subtraction in python
v1 = np.array([1,2,3])
v2 = np.array([5,6,7])
result = v1 + v2
result
```

```python
# Scalar Multiplication
result * 3
```

```python
# Dot Product of vector
dot = np.dot(v1,v2)
print(dot)
```
