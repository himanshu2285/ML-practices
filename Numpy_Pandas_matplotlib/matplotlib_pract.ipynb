# Visualization with matplotlib

```python

import matplotlib.pyplot as plt
import numpy as np
```

### 1. Line Chart:
It is helpful for visualizing Continous Data

```python
x = np.array([2022, 2023, 2024, 2025])
y1 = np.array([50, 40, 65, 30])
y2 = np.array([30, 60, 40, 50])
y3 = np.array([40, 30, 50, 40])

line_style = dict(marker='.',       # packing
                  markersize=20,
                  markerfacecolor="#0fe9ec",
                  markeredgecolor="#0fe9ec",
                  linestyle='solid',  # dashed, dotted, dashdot, solid, None
                  linewidth=2,)

plt.plot(x, y1, color="#2030ea",**line_style)       # unpacking
plt.plot(x, y2, color="#23de39", **line_style)
plt.plot(x, y3, color="#c61d10", **line_style)
plt.xlabel("Years")
plt.ylabel("Students")
plt.title("Student Line Graph",
                fontsize=20,
                fontweight="bold",
                color="#da0404",)
plt.show()
```

Grid() = Helps make plot easier to read by adding reference lines.

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([10,20,15,25,30])

plt.grid()  # x,y,both

plt.plot(x,y)
plt.show()
```

### 2. Bar chart:
It is helpful in categorical data visualization

```python
import numpy as  np
import matplotlib.pyplot as plt

category = np.array(["Apple", "Banana","Cherry","Grapes","Mango"])
values = np.array([4,2,7,6,5])

plt.bar(category, values, color="#f66c4d", edgecolor="#c70039", linewidth=2)
# plt.barh(category, values, color="#f66c4d", edgecolor="#c70039", linewidth=2)   # For horizontal bar graph

plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.title("Fruit Quantity")

plt.show()
```

### 3. Pie Chart
Circular chart divided into slices to show percentage of the total. <br>
Good for visualizing distribution among the categories.

```python
animal_category = np.array(["Mammals", "Birds", "Reptiles", "Amphibians", "Fish"])
animal_values = np.array([15, 10, 8, 5, 12])

plt.pie(animal_values, labels=animal_category,
        autopct="%1.1f%%", 
        startangle=90,
        explode=[0.1, 0, 0, 0, 0],
        shadow=True
        )
plt.title("Animal Categories")

plt.show()
```

### 4. Scatter Plot:
Shows the relationship between two variables <br>
Helps to identify a correlation (+, -, None) <br>
Example: Study hours VS Test Score

```python
x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]) # Hours studied
y1= np.array([55,60,65, 62, 68,70, 75,78,82,85, 87]) # Grades

x2 = np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8]) # Hours studied
y2 = np.array([50,58, 65, 70, 72, 78, 83, 88,92, 95, 97]) # Grades

plt.scatter(x1, y1, color="skyblue",
            alpha = 0.6,
            s = 200,
            label="Group A")
plt.scatter(x2, y2, color="lightcoral",
            alpha = 0.6,
            s = 200,
            label="Group B")

plt.xlabel("Hours Studied")
plt.ylabel("Grades")
plt.title("Study Hours vs Grades")

plt.legend()
plt.show()
```
