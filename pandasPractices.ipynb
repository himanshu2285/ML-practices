Converting JSON data into Tabular Form

```python
olympic_list = {"countries":['London','Beijing','Sydney','Athens','India'],
                'year':[2000,2001,2002,2003,np.nan],
                'TotalCountry':[205,204,201,200,np.nan]}
olympic_list
```

np.nan is for missing data, i.e. nan(not a number)

```python
olympic_list_df = pd.DataFrame(olympic_list)
olympic_list_df
```

### Handling Missing Values

DataFrame.dropna() is used to drop entire row that have missing numbers

```python
olympic_list_df.dropna(how="any")
```

```python
olympic_list_df[olympic_list_df['year']>2000]
```

```python
olympic_list_df.columns
```
