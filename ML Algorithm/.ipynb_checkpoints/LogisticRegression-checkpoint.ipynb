# Logistic Regression Practice

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
%matplotlib inline
```

```python
digits = load_digits()
```

```python
print(f"Image Data shape {digits.data.shape}")
print(f"Label Data shape {digits.target.shape}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.23, random_state=2)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

### Trainig Model with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegr = LogisticRegression(max_iter=1000)
logisticRegr.fit(x_train, y_train)
```

```python
# Returns a Numpy Array
# Predict for one observation

print(logisticRegr.predict(x_test[0].reshape(1, -1)))
```

```python
logisticRegr.predict(x_test[0:10])
```

```python
predictions = logisticRegr.predict(x_test)
predictions
```

### Checking Model Accuracy

```python
# score() is used to check the accuracy of the model
score = logisticRegr.score(x_test, y_test)
score
```

### Model Analysis
More the numbers we have diagnol, more the model is accurate

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, predictions)
cm
```

```python
plt.figure(figsize=(9, 9))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
