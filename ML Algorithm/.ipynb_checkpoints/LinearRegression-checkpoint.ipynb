# Use Case Implementation of Linear Regression

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
componies = pd.read_csv("../datasets/csv/1000_Companies.csv")
x = componies.iloc[:, :-1].values
y = componies.iloc[:, 4].values

componies.head()
```

```python
x
```

```python
y
```

```python
# Data Visualization
sns.heatmap(componies.corr(numeric_only=True))
```

### Encoding Categorical Data

```python
# Encoding Categorical Data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
labelencoder = LabelEncoder()
x[:, 3] = labelencoder.fit_transform(x[:, 3])
```

```python
# onehotencoder = OneHotEncoder(categorical_features = [3])
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [3])
    ],
    remainder='passthrough'
)
x = ct.fit_transform(x)
```

```python
x
```

### Avoiding Dummy Variable Trap

```python
x = x[:, 1:]
```

### Splitting the Data Into Train and Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### Fitting Multiple Linear Regression Model to training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

### Predicting the Test Set Results

```python
y_pred = regressor.predict(x_test)
y_pred
```

### Calculating coefficient and intercept

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```
